```python
import os 
import pandas as pd
import random
```

```python
normal_directory = "Normal"

# Create a list to store normal DataFrames
normal_dataframes = []

# Loop through normal traffic files
for filename in os.listdir(normal_directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(normal_directory, filename)
        df = pd.read_csv(file_path)
        # Randomly select 30% of the file
        sample_size = int(len(df) * 0.2)
        random_indices = random.sample(range(len(df)), sample_size)
        df = df.iloc[random_indices]
        # Add label column
        df['label'] = 'normal'
        normal_dataframes.append(df)

# Concatenate all the sampled DataFrames
normal_df = pd.concat(normal_dataframes, ignore_index=True)
```

```python
normal_df
```

```python
attack_directory = "Attack"

# Create a list to store attack DataFrames
attack_dataframes = []

# Loop through attack traffic files
for filename in os.listdir(attack_directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(attack_directory, filename)
        df = pd.read_csv(file_path)
        # Randomly select 80% of the file
        sample_size = int(len(df) * 0.4)
        random_indices = random.sample(range(len(df)), sample_size)
        df = df.iloc[random_indices]
        # Add label column
        df['label'] = 'attack'
        attack_dataframes.append(df)

# Concatenate all the sampled DataFrames
attack_df = pd.concat(attack_dataframes, ignore_index=True)
```

```python
attack_df
```

```python
combined_df = pd.concat([normal_df, attack_df], ignore_index=True)
```

```python
combined_df
```

```python
combined_df.to_csv('final.csv')
```
